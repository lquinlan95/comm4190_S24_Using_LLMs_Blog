## Multimedia Matchmaker (Part I)

I have been playing with ChatGPT for some time now, but I have never taken the opportunity to explore the GPT Builder function that comes with the OpenAI system. The GPT builder enables you to build GPT's that are specifically tailored to your needs. For example, one of my friends trained a GPT to identify hate speech.

I will design a GPT for a simpler task: recommend media for me to try.

### Initial set-up

GPT Builder is pretty intuitive to use, especially considering ChatGPT's ability to communicate and understand language so clearly. The initial set-up only required casual instruction on my part, with brief questions from ChatGPT.

Below is the initial page that you see when you go into the GPT builder: 

On the left-hand side, I input the initial instructions:

"I would like to create a multimedia recommendation GPT. The system should be able to take input from the user about their mood, their interests, and previous media that they have consumed and recommend new media to them. Recommendations can include the realms of books, tv shows, movies, podcasts, or songs.

Example prompts might include:
1) "I just finished book x, can you recommend a tv show that explores similar themes?"
2) I am reading book x, can you recommend some songs for a soundtrack while I read that fit that vibe?"
3) I am feeling romantic, can you recommend a good tv show about love?
4) I want to explore time travel more
5) What are some good horror books that aren't Steven King
6) Can you recommend me a podcast about similar topics as tv show x?"

This enabled me to create the initial shape of the GPT. In the builder, there is a create tab, as well as a configure tab. The Configure tab is where we can go to specifically change the attributes of the GPT. After this initial request, the Configure tab showed:

### Tweaking the configuration

To further refine the configuration of the GPT, the Builder asked as series of follow up questions:

- **ChatGPT:** "First off, how detailed do you want the recommendations to be? Should they include a brief description of the media, why it matches your request, and potentially where you can find them?"
- **Me:** "The recommendations should include a brief description of the media and why it matches the request."
- **GPT Builder:** "Next, how should Media Matchmaker approach situations where the details provided by the user are vague or incomplete? Would you prefer it to ask for more specific information or make assumptions based on what's provided?"
- **Me:** "It should ask follow up questions and probe deeper if the details provided are vague. Like if someone says " I want to watch something funny," Media Matchmaker can ask, "Are you looking for a rom com, a buddy comedy, a dark comedy," etc"
- **GPT Builder:** "How do you want Media Matchmaker to express its personality? Should it be enthusiastic and passionate about media, professional and straightforward, or maybe have a quirky sense of humor?"
- **Me:** "It should be more straightforward, like an expert explaining a topic. It should almost sound like if you went to a librarian to ask for a book recommendation"

After the configurations, the Configure tab showed how the Media Matchmaker GPT was being prompted:

"You're **Media Matchmaker**, an expert in multimedia recommendations, akin to a knowledgeable librarian. Your role is to suggest new media—books, TV shows, movies, podcasts, and songs—based on a user's mood, interests, and previously consumed media. You provide tailored suggestions with a brief description and rationale, emphasizing a straightforward, professional approach in your explanations. When encountering vague requests, you engage with follow-up questions to gain clarity, ensuring your recommendations are precise and valuable. Your demeanor is informative, focusing on delivering insightful recommendations without overwhelming users, facilitating an enriching media discovery experience."

GPT Builder was a super intuitive tool to set up and explore. The natural language of the system makes translating my intentions into code that is useful for tailoring the GPT system incredibly efficient. Additionally, the Configure tab provides some input on system prompting.
